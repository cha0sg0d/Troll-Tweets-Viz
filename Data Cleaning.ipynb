```python
from sklearn.feature_extraction import text
from datetime import date
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
frames = []
for i in range(1,11):
    a = "IRAhandle_tweets_" + str(i) + ".csv"
    frames.append(pd.read_csv(a,usecols=["account_type", "publish_date", "content"]))
```

```python
df1 = pd.concat(frames)
```

```python
df1['publish_date'] = pd.to_datetime(df1['publish_date'])
```

```python
df1.shape
```

```python
left = df1[df1.account_type == "Left"]
right = df1[df1.account_type == "Right"]
```

```python
print(left.shape)
print(right.shape)
```

```python
my_stop_words = text.ENGLISH_STOP_WORDS.union(["https","http","ha","he","rt","gt","amp"])
```

```python
# General flow

# Union data

# Store left, right columns

# Timestamp.replace(self, year=None, month=None, day=None, hour=None, minute=None, second=None, microsecond=None, nanosecond=None, tzinfo=<class 'object'>, fold=0)Â¶
```

```python
first = pd.date_range(start='7/1/2015', end='12/31/2017', freq="MS")
last = pd.date_range(start='7/1/2015', end='12/31/2017', freq="M")
```

```python
# for all data left by month
left_results = []
for start,end in zip(first,last):
    curr_tweets = left[(left['publish_date'] > start) & (left['publish_date'] < end.replace(hour=23))]
    if not curr_tweets.empty:
        vectorizer = CountVectorizer(strip_accents = "unicode",lowercase = True, stop_words = my_stop_words,max_features = 10,ngram_range = (2,2))
        # tokenize and build vocab
        vectorizer.fit(curr_tweets['content'])
        # summarize
        #print(vectorizer.vocabulary_)
        # encode document
        vector = vectorizer.transform(curr_tweets['content'])
        # summarize encoded vector
        a = np.sum(vector.toarray(),axis=0)
        
        for key, value in vectorizer.vocabulary_.items():
            final_dict = {}
            final_dict["date"] = start.strftime("%m/%d/%Y") 
            final_dict["word"] = key
            final_dict["value"] = a[value]
            print(final_dict,"\n")
            left_results.append(final_dict)
            
```

```python
# for all data right by month
right_results = []
for start,end in zip(first,last):
    curr_tweets = right[(right['publish_date'] > start) & (right['publish_date'] < end.replace(hour=23))]
    if not curr_tweets.empty:
        vectorizer = CountVectorizer(strip_accents = "unicode",lowercase = True, stop_words = my_stop_words, max_features = 10, ngram_range = (2,2))
        # tokenize and build vocab
        vectorizer.fit(curr_tweets['content'])
        # summarize
        #print(vectorizer.vocabulary_)
        # encode document
        vector = vectorizer.transform(curr_tweets['content'])
        # summarize encoded vector
        a = np.sum(vector.toarray(),axis=0)
        for key, value in vectorizer.vocabulary_.items():
            final_dict = {}
            final_dict["date"] = start.strftime("%m/%d/%Y") 
            final_dict["word"] = key
            final_dict["value"] = a[value]
            right_results.append(final_dict) 
```

```python
finished = {"left": left_results, "right": right_results}
```
